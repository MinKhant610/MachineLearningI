```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt 

def main():
    img = cv.imread('images/4.2.07.tiff')
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    
    identity = np.array( ( [0, 0, 0], [0, 1, 0], [0, 0, 0]) , np.float32)
    ed = np.array( ( [0, -1, 0], [-1, 4, -1], [0, -1, 0]) , np.float32)
    sh = np.array( ( [0, -1, 0], [-1, 5, -1], [0, -1, 0]) , np.float32)
    bo = np.ones((11, 11), np.float32)/ 121
    
    iden = cv.filter2D(img, -1, identity)
    edge = cv.filter2D(img, -1, ed)
    sharp = cv.filter2D(img, -1, sh)
    box = cv.filter2D(img, -1, bo)
    
    images = [iden, edge, sharp, box]
    titles = ['identity', 'edge', 'sharp', 'box']
    
    for i in range(len(images)):
        plt.subplot(2, 2, i+1)
        plt.title(titles[i])
        plt.axis('off')
        plt.imshow(images[i])
    plt.show()
    

if __name__ == '__main__':
    main()
```

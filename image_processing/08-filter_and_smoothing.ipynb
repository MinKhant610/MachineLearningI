```python
import cv2 as cv
import matplotlib.pyplot as plt 
import numpy as np 
import random 

def main():
    img = cv.imread('images/4.2.02.tiff')
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    
    row, col, ch = img.shape
    p = 0.05 
    
    output = np.zeros(img.shape, np.uint8)
    
    for i in range(row):
        for j in range(col):
            r = random.random() # float between 1 and 0
            
            if r < p/2 : 
                # pepper and sprinkled
                output[i][j] = [0, 0, 0]
            elif r < p :
                # salt and sprinkled
                output[i][j] = [255, 255, 255]
            else :
                output[i][j] = img[i][j]
    
    denoise = cv.medianBlur(output, 3)
    blur = cv.GaussianBlur(output, (5, 5), 0)
    
    
    images = [img, output, denoise, blur]
    titles = ['original', 'noise', 'median', 'gaussian']
    
    for i in range(len(images)):
        plt.subplot(2, 2, i+1)
        plt.title(titles[i])
        plt.axis('off')
        plt.imshow(images[i])
    plt.show()
    

if __name__ == '__main__':
    main()
```

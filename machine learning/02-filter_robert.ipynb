```python
from skimage import io, restoration 
from matplotlib import pyplot as plt 
from skimage.filters.rank import entropy
from skimage.morphology import disk

img = io.imread('images/balloon.jpg', as_gray=True)
entropy_img = entropy(img, disk(3))

from skimage.filters import try_all_threshold
from skimage.filters import threshold_otsu

thresh = threshold_otsu(entropy_img)

binary = entropy_img <= thresh

images = [img ,entropy_img, binary]
titles = ['original','entropy', 'binary']

for i in range(len(images)):
    plt.subplot(1, 3, i+1)
    plt.title(titles[i])
    plt.axis('off')
    plt.imshow(images[i], cmap='gray')
plt.show()
```

```python
from skimage import data 

img = data.coins()
from skimage.filters import roberts, sobel, scharr, prewitt
import matplotlib.pyplot as plt 

edge_robert = roberts(img)
edge_soble = sobel(img)
edge_scharr = scharr(img)
edge_prewitt = prewitt(img)

images = [edge_robert, edge_soble, edge_scharr, edge_prewitt]
titles = ['edge_robert', 'edge_soble', 'edge_scharr', 'edge_prewitt']
plt.figure(figsize=(20, 10))

for i in range(len(images)):    
    plt.subplot(2, 2, i+1)
    plt.title(titles[i])
    plt.axis('off')
    plt.imshow(images[i], cmap='gray')
plt.show()
```

```python
from skimage.feature import canny
edge_canny = canny(img, sigma=5)
plt.imshow(edge_canny, cmap='gray')
```

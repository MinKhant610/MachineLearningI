# sum of image matrix or elements

```python
import cv2 as cv 

def main():
    img = cv.imread('images/T.jpg', 0)
    
    print(cv.sumElems(img))

if __name__ == '__main__':
    main()
```

# Binarization

```python
import cv2 as cv 
import matplotlib.pyplot as plt 

def main():
    img = cv.imread ('images/gray21.512.tiff')
    img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    
    plt.subplot(1, 2, 1)
    plt.title('Gray')
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img, cmap='gray')
    
    plt.subplot(1, 2, 2)
    plt.title('Histogram')
    plt.hist(img.ravel(), 255, [0, 256])
    plt.xlim(xmin=0, xmax=256)
    plt.show()
    
    th = 100
    max_val = 1 
    
    ret, output1 = cv.threshold(img, th, max_val, cv.THRESH_BINARY)
    ret, output2 = cv.threshold(img, th, max_val, cv.THRESH_BINARY_INV)
    
    print (output1)
    
    titles = ['binary', 'binary_inverse']
    images = [output1, output2]
    
    for i in range(len(images)):
        plt.subplot(1, 2, i+1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
        
    plt.show()

if __name__ == '__main__':
    main()
```

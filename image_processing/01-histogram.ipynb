```python
import cv2 as cv
import matplotlib.pyplot as plt 


def main():
    img = cv.imread('images/T.jpg')
    
    plt.subplot(1, 2, 1)
    plt.title('T shape')
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img, cmap='gray')
    
    plt.subplot(1, 2, 2)
    plt.title('Histogram')
    plt.hist(img.ravel(), 256, [0, 256])
    plt.xlim(xmin=0, xmax=256)
    plt.show()
    


if __name__ == '__main__':
    main()
```

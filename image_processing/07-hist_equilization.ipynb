```python
import cv2 as cv
import matplotlib.pyplot as plt


def main():
    img = cv.imread('images/industrial.jpg', 0)
    out1 = cv.equalizeHist(img)
    
    clahe = cv.createCLAHE(clipLimit=44.0, tileGridSize=(8, 8))
    out2 = clahe.apply(img)

    images = [img, out1, out2]
    titles = ['original', 'equalizeHist', 'CLAHE']

    plt.figure(figsize=(20, 10))

    for i in range(len(images)):
        plt.subplot(1, 3, i+1)
        plt.title(titles[i])
        plt.axis('off')
        plt.imshow(images[i], cmap='gray')
    plt.show()

    plt.figure(figsize=(20, 10))
    
    for i in range(len(images)):
        plt.subplot(1, 3, i+1)
        plt.title(titles[i])
        plt.hist(images[i].ravel(), 256, [0, 255])
        plt.xlim(xmin=0, xmax=256)
    plt.show()


if __name__ == '__main__':
    main()
```

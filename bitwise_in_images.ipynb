```python
import cv2 as cv 
import matplotlib.pyplot as plt 

def main():
    img1 = cv.imread('images/lena_color_512.tif')
    img2 = cv.imread('images/4.2.01.tiff')
    
    img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
    img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
    
    img3 = cv.bitwise_not(img1)
    img4 = cv.bitwise_or(img1, img2)
    img5 = cv.bitwise_and(img1, img2)
    img6 = cv.bitwise_xor(img1, img2) 
    
    images = [img1, img2, img3, img4, img5, img6]
    titles = ['lena_color', '4.2', 'lena(not)', 'lena or 4.2', 'lena and 4.2', 'lena xor 4.2']
    
    for i in range(len(images)):
        plt.subplot(2, 3, i+1)
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
        plt.imshow(images[i])
    plt.show()

if __name__ == '__main__':
    main()
```

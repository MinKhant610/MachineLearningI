```python
import cv2 as cv
import matplotlib.pyplot as plt 

def main():
    img1 = cv.imread('images/lena_color_512.tif')
    img2 = cv.imread('images/house.tif')
    
    img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
    img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
    
    img3 = img1 * 1.5
    img4 = img2 / 1.5 
    
    add = img1 + img2 
    sub1 = img1 - img2
    sub2 = img2 - img1
    
    mult = img1 * img2
    lena50 = img1 + 50
    lena_50 = img1 - 50
    
    fifty = 50 - img1
    division = img1/img2 
    
    images = [img1, img2, img3, img4, add, sub1, sub2, mult, lena50, lena_50, fifty, division]
    titles = ['lena', 'house', 'lena * 1.5', 'house * 1.5',  'lena + house', 'lena - house', 'house- lena', 'lena * house', 'lena + 50', 'lena  - 50', '50 - lena', 'lena / house']
    
    for i in range(len(images)):
        plt.subplot(3, 4, i+1)
        plt.xticks([])
        plt.yticks([])
        plt.title(titles[i])
        plt.imshow(images[i])
    plt.show()
    
    

if __name__ == '__main__':
    main()
```

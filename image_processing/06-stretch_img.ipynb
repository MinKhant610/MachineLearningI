```python
import cv2 as cv 
import matplotlib.pyplot as plt 
import numpy as np

def main():
    img = cv.imread('images/industrial.jpg')
    original = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
        
    min_val = np.min(gray)
    max_val = np.max(gray)
    
    #stretch_img
    stretch_img = (255 / (max_val - min_val) * (gray - min_val))
    stretch_img = np.uint8(stretch_img)
    
    images = [original, stretch_img]
    titles = ['original', 'stretch_img']
    
    plt.figure(figsize=(20, 10))
    
    for i in range(len(images)):
        plt.subplot(1, 2, i+1)
        plt.title(titles[i])
        plt.axis('off')
        plt.imshow(images[i], cmap='gray')
    plt.show()
    
    plt.figure(figsize=(20, 10))
    
    plt.subplot(1, 2, 1)
    plt.title('original')
    plt.hist(original.ravel(), 256, [0, 256])
    plt.xlim(xmin=0, xmax=256)
    
    plt.subplot(1, 2, 2)
    plt.title('stretch_img')
    plt.hist(stretch_img.ravel(), 256, [0, 256])
    plt.xlim(xmin=0, xmax=256)
    
    plt.show()
    

if __name__ == '__main__':
    main()
```

```python
import cv2 as cv 
import matplotlib.pyplot as plt 

def main():
    img1 = cv.imread('images/lake_color.tiff')
    
    img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
    img2 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
    
    img3 = 255 - img1
    img4 = 255 - img2 
    
    images = [img1, img2, img3, img4]
    titles = ['lake_color', 'lake_gray', '255-lake_color(neg)', '255 - gray(neg)']
    
    for i in range(len(images)):
        plt.subplot(2, 2, i+1)
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
        plt.imshow(images[i], cmap='gray')
        
    plt.show()

if __name__ == '__main__':
    main()
```

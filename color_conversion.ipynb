# Color Conversion Flag in OpenCV

```python
import cv2 as cv 
import matplotlib.pyplot as plt 

def main():
    j = 0
    for filename in dir(cv):
        if filename.startswith('COLOR_'):
            print(filename)
            j+=1
    print ('There are ' + str(j+1) + 'color conversion flags')

if __name__ == '__main__':
    main()
```

```python
img = cv.imread('images/lena_color_512.tif', 1)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('Lena Color')
plt.xticks([])
plt.yticks([])
plt.show()
```

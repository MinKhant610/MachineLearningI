```python
from skimage import io
from matplotlib import pyplot as plt 

img = io.imread('images/boat.512.tiff', as_gray=True)

from skimage.transform import resize, rescale, downscale_local_mean


resize_img = resize(img, (200, 200), anti_aliasing=True)
rescale_img = rescale(img, 1.0/2.0, anti_aliasing=True)
down_img = downscale_local_mean(img,(1,1))

plt.imshow(down_img, cmap='gray')
```

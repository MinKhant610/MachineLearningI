```python
import cv2 as cv 
import matplotlib.pyplot as plt 

def main():
    img1 = cv.imread('images/lena_color_512.tif')
    img2 = cv.imread('images/house.tif')
    
    lena = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
    house = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
    
    # alpha + beta must be 1 
    alpha = 0.6 
    beta = 0.4
    gamma = 0 
    
    output1 = lena + house
    output2 = cv.addWeighted(lena, alpha, house, beta, gamma)
    
    images = [lena, house, output1, output2]
    titles = ['lena', 'house', 'lena + house', 'lena + house (blending)']
    
    for i in range(len(images)):
        plt.subplot(2, 2, i+1)
        plt.xticks([])
        plt.yticks([])
        plt.title(titles[i])
        plt.imshow(images[i])
    plt.show()

if __name__ == '__main__':
    main()
```

```python
import cv2 as cv 
import matplotlib.pyplot as plt 

def main():
    img = cv.imread('images/7.1.10.tiff')
    
    img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    
    th = 0
    max_val = 1
    
    ret, output1 = cv.threshold(img, th, max_val, cv.THRESH_BINARY + cv.THRESH_OTSU)
    ret, output2 = cv.threshold(img, th, max_val, cv.THRESH_BINARY_INV + cv.THRESH_OTSU)
    ret, output3 = cv.threshold(img, th, max_val, cv.THRESH_TOZERO + cv.THRESH_OTSU)
    ret, output4 = cv.threshold(img, th, max_val, cv.THRESH_BINARY + cv.THRESH_OTSU)
    ret, output5 = cv.threshold(img, th, max_val, cv.THRESH_TRUNC + cv.THRESH_OTSU)
    
    images = [img, output1, output2, output3, output4, output5]
    titles = ['original', 'bin', 'bin_inv', 'to zero', 'to zero inv', 'trunc']
    
    for i in range(len(images)):
        plt.subplot(2, 3, i+1)
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
        plt.imshow(images[i], cmap='gray')
    plt.show()

if __name__ == '__main__':
    main()
```

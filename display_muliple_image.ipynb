```python
import cv2 as cv 
import matplotlib.pyplot as plt 

def main():
    img1 = cv.imread('images/lena_color_512.tif')
    img2 = cv.imread('images/lake.tif')
    
    img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
    img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
    
    plt.subplot(1, 2, 1)
    plt.title('Lena')
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img1)
    plt.show()
    
    plt.subplot(1, 2, 2)
    plt.title('lake')
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img2)
    plt.show()

if __name__ == '__main__':
    main()
```

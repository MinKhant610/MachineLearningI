```python
import cv2 as cv
import matplotlib.pyplot as plt 

def main():
    img = cv.imread('images/football.jpg')
    img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    
    w, h = img.shape
    total = w * h
    print ('total :', total)
    
    th = 200
    max_val = 1
    
    ret, out = cv.threshold(img, th, max_val, cv.THRESH_BINARY)
    
    white_size = cv.sumElems(out)
    
    print ('white size :', white_size[0])
    
    object_size = white_size[0]
    print ('obj size', object_size)
    
    images = [img, out]
    titles = ['football', 'bin']
    
    for i in range(len(images)):
        plt.subplot(1, 2, i+1)
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
        plt.imshow(images[i], cmap='gray')
    plt.show()
    
    
if __name__ == '__main__':
    main()
```
